Having a dry hitting streak would be I very negative hit on your BPAB. These players contain non of the huge name players like Mike Trout or Bryce Harper which is interesting considering their large contracts and hitting ability, this could be due to the team around them and the positions they are hitting. If you are a lead of hitter then you will have very few chances to move other people on the bases. If you are hitting 4th in the line up you are going to have much more chances at moving people. On the same note even if you are hitting fourth in the line up, but your teammates are never getting on bases then you will have no chance to move players.
