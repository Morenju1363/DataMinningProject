```python
import pandas as pd
import numpy as np
from pybaseball.lahman import *
from pybaseball import bwar_bat
from pybaseball import *

# get WAR stats from baseball reference 
data = bwar_bat() 
data.head()
# batting stats by year, regular season
bat = batting()
bat.head()
#On Base Percentage = HITS + Walks + Hts by Pitch / At Bats + Walks + Hits by Pitch + Sacrafice Flies
bat['OBP'] = (bat['H'] + bat['BB'] + bat['HBP'])/(bat['AB'] + bat['BB'] + bat['HBP'] + bat['SF'])
#Batting Average = Hits / At Bats 
bat['BA'] = (bat['H']/bat['AB'])
#Plate Appearance = At Bats + Walks + Sacrafice Flies + Stolen Base
bat['PA'] = (bat['AB']+bat['BB']+bat['SF']+bat['SB'])

bat = bat.loc[(bat["yearID"] == 2017),["playerID","yearID","OBP",'BA','PA']]
bat.head(10)
data = data.loc[(data["year_ID"] == 2017),["player_ID","year_ID","WAR",'salary']]
data.columns = ['playerID','yearID','WAR','salary']
data.head(100)
data.dropna()
bat.dropna()
final = pd.merge(data[['playerID','yearID','WAR','salary']],bat[['playerID','yearID','OBP','BA','PA']],on=['playerID','yearID'],how = 'inner')
final = final.dropna()
final = final.loc[(final["PA"]>=10)]
final.head(100)
```
